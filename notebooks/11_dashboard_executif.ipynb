# Rapport de conclusion
## Resumé executif

Cette analyse porte sur le parc immobilier communal afin d'identifier les leviers d'optimisation énergétique. Les données révèlent une forte corrélation entre les chutes de température hivernales et les pics de consommation, particulièrement pour le gaz. Bien que la surface soit un facteur déterminant, l'hétérogénéité des consommations au sein d'une même classe DPE suggère des gisements d'économies liés aux comportements des usagers et à la vétusté des équipements. Le plan d'action préconise une priorité de rénovation sur les bâtiments de classes F et G du secteur résidentiel.
